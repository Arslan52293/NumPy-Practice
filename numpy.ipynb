                     #JOINING NUMPY ARRAY

#Vstack

```python
a1 = np.array([1,2,3,4])
a2 = np.array([5,6,7,8])
```

```python
np.vstack((a1,a2))
```

#Hstack horizontal stack

```python
np.hstack((a1,a2))
```

#Col stack

```python
np.column_stack((a1,a2))
```

#Intersection and diff

```python
n1 = np.array([10,20,30,40,50])
n2 = np.array([20,50,80,60,30])
```

```python
np.intersect1d(n1,n2)
```

```python
np.setdiff1d(n1,n2)
```

            # ARRAY MATHEMATICS

```python
x1 = np.array([10,20])
x2 = np.array([30,40])
```

```python
np.sum([x1,x2])
```

```python
np.sum([x1,x2],axis=0)                       #  col val
```

```python
np.sum([x1,x2],axis=1)                      #row val
```
